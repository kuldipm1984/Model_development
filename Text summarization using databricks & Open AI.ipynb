```python
from pyspark.sql.functions import collect_list, concat_ws

text_df = df.select(
    concat_ws(" ", collect_list("value")).alias("document_text")
)

document_text = text_df.first()["document_text"]
print(document_text[:500])
```

```python
OPENAI_ENDPOINT = "https://XXXXXXXXXXXXX.openai.azure.com/"
OPENAI_KEY = "XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX"
from openai import AzureOpenAI

EMBEDDING_MODEL = "agnetic_model"

client = AzureOpenAI(
    api_key=OPENAI_KEY,
    azure_endpoint=OPENAI_ENDPOINT,
    api_version="2024-12-01-preview"
)
```

```python
def summarize_text(text):
    prompt = f"""
    Summarize the following document in 5 bullet points.
    Be concise and factual.

    DOCUMENT:
    {text}
    """

    response = client.chat.completions.create(
        model="agnetic_model",  
        messages=[
            {"role": "system", "content": "You are an expert summarizer."},
            {"role": "user", "content": prompt}
        ],
        temperature=0.2
    )

    return response.choices[0].message.content
```

```python
summary = summarize_text(document_text)
print(summary)
```
